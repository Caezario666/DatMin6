```python
import pandas as pd
```

```python
file_path = 'C:/Semester 5/Data Mining/Pert 6/Data_1.xlsx'
data = pd.read_excel(file_path)
```

```python
data.head(), data.info()
```

```python
data = data.drop(columns=['Unnamed: 12'])
```

```python
data.columns = data.columns.str.strip()
```

```python
data['USIA'].unique(), data['JARAK KELAHIRAN'].unique(), data['PE/Non PE'].unique()
```

```python
import re
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['USIA'] = data['USIA'].astype(str).apply(lambda x: re.sub(r'\D', '', x)).replace('', None).astype(float)
```

```python
data['JARAK KELAHIRAN'] = data['JARAK KELAHIRAN'].str.strip().replace({
    'anak pertama': 'anak pertama',
    '< 2 tahun': '< 2 tahun',
    '> 2 tahun ': '> 2 tahun',
    '> 2 tahun': '> 2 tahun'
})
```

```python
data['PE/Non PE'] = data['PE/Non PE'].replace({
    'Non PE': 'Non PE',
    'PE': 'PE',
    'PEB': 'PE',
    'Eklamsia': 'PE',
    'PE gemelli': 'PE',
    'PEB impending eklampsi': 'PE',
    'PE ': 'PE'
})
```

```python
data['USIA'] = data['USIA'].fillna(data['USIA'].median())
```

```python
data['JARAK KELAHIRAN'] = data['JARAK KELAHIRAN'].fillna(data['JARAK KELAHIRAN'].mode()[0])
```

```python
categorical_columns = data.select_dtypes(include=['object']).columns
```

```python
label_encoders = {col: LabelEncoder() for col in categorical_columns}
```

```python
for col, le in label_encoders.items():
    data[col] = le.fit_transform(data[col])
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
```

```python
X = data.drop(columns=['NO', 'NAMA', 'PE/Non PE'])  # Exclude identifiers and target variable
y = data['PE/Non PE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
nb_model = GaussianNB()
knn_model = KNeighborsClassifier()
dt_model = DecisionTreeClassifier(random_state=42)
```

```python
nb_model.fit(X_train, y_train)
knn_model.fit(X_train, y_train)
dt_model.fit(X_train, y_train)
```

```python
y_pred_nb = nb_model.predict(X_test)
y_pred_knn = knn_model.predict(X_test)
y_pred_dt = dt_model.predict(X_test)
```

```python
cm_nb = confusion_matrix(y_test, y_pred_nb)
cm_knn = confusion_matrix(y_test, y_pred_knn)
cm_dt = confusion_matrix(y_test, y_pred_dt)
```
